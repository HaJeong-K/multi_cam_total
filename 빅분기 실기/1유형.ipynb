```python
# 문제 12
# hp 변수에 대해 데이터표준화(Z-score)를 진행하고 이상치의 수를 구하시오.
# (단, 이상치는 Z값이 1.5를 초과하거나 -1.5 미만인 값이다)
df = pd.read_csv("mtcars.csv")
```

```python
# 틀림
# Z = (x-평균) / 표준편차

std = df['hp'].std()
mean = df['hp'].mean()
df['Z'] = (df['hp'] - mean) / std
cond1 = (df['Z'] > 1.5)
cond2 = (df['Z'] < -1.5)
print(len(df[cond1 | cond2]))
```

```python
# 문제 13
# mpg 컬럼을 최소최대 Scaling을 진행한 후 0.7보다 큰 값을 가지는 레코드 수를 구하라.
df = pd.read_csv("mtcars.csv")
```

```python
# 틀림
# 공식 : (x - min) / (max - min)

min_mpg = df['mpg'].min()
max_mpg = df['mpg'].max()
df['mpg'] = (df['mpg'] - min_mpg) / (max_mpg - min_mpg)
cond1 = (df['mpg'] > 0.7)
print(len(df[cond1]))
```

```python
# 문제 14
# wt컬럼에 대해 상자그림 기준으로 이상치의 개수를 구하시오.
df = pd.read_csv("mtcars.csv")
```

```python
# 틀림
Q1 = df['wt'].quantile(0.25)
Q3 = df['wt'].quantile(0.75)
IQR = Q3 - Q1

upper = Q3 + (IQR * 1.5)
lower = Q1 - (IQR * 1.5)

cond1 = (df['wt'] < lower)
cond2 = (df['wt'] > upper)
print(len(df[cond1]) + len(df[cond2]))
```
